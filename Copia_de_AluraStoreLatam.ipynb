### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

print("Ingreso Tienda 1:", ingreso_tienda)
print("Ingreso Tienda 2:", ingreso_tienda2)
print("Ingreso Tienda 3:", ingreso_tienda3)
print("Ingreso Tienda 4:", ingreso_tienda4)
```

# 2. Ventas por categoría

```python
categoria_tienda1 = tienda["Categoría del Producto"].value_counts()
categoria_tienda2 = tienda2["Categoría del Producto"].value_counts()
categoria_tienda3 = tienda3["Categoría del Producto"].value_counts()
categoria_tienda4 = tienda4["Categoría del Producto"].value_counts()

print("Tienda 1")
print(categoria_tienda1)

print("\nTienda 2")
print(categoria_tienda2)

print("\nTienda 3")
print(categoria_tienda3)

print("\nTienda 4")
print(categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
promedio_tienda1 = tienda["Calificación"].mean()
promedio_tienda2 = tienda2["Calificación"].mean()
promedio_tienda3 = tienda3["Calificación"].mean()
promedio_tienda4 = tienda4["Calificación"].mean()

print("Calificación promedio Tienda 1:", promedio_tienda1)
print("Calificación promedio Tienda 2:", promedio_tienda2)
print("Calificación promedio Tienda 3:", promedio_tienda3)
print("Calificación promedio Tienda 4:", promedio_tienda4)
```

```python
import matplotlib.pyplot as plt

# Datos
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
calificaciones = [promedio_tienda1, promedio_tienda2, promedio_tienda3, promedio_tienda4]

plt.figure()

plt.pie(
    calificaciones,
    labels=tiendas,
    autopct='%1.2f%%',
    startangle=90
)

plt.title("Distribución de la calificación promedio por tienda")

plt.show()
```

# 4. Productos más y menos vendidos

```python
ventas_tienda1 = tienda["Producto"].value_counts()
ventas_tienda2 = tienda2["Producto"].value_counts()
ventas_tienda3 = tienda3["Producto"].value_counts()
ventas_tienda4 = tienda4["Producto"].value_counts()

print("Tienda 1")
print("Más vendido:", ventas_tienda1.idxmax())
print("Menos vendido:", ventas_tienda1.idxmin())

print("\nTienda 2")
print("Más vendido:", ventas_tienda2.idxmax())
print("Menos vendido:", ventas_tienda2.idxmin())

print("\nTienda 3")
print("Más vendido:", ventas_tienda3.idxmax())
print("Menos vendido:", ventas_tienda3.idxmin())

print("\nTienda 4")
print("Más vendido:", ventas_tienda4.idxmax())
print("Menos vendido:", ventas_tienda4.idxmin())
```

```python
import matplotlib.pyplot as plt

productos = [
    ventas_tienda1.idxmax(),
    ventas_tienda1.idxmin()
]

cantidades = [
    ventas_tienda1.max(),
    ventas_tienda1.min()
]

plt.figure()
plt.bar(productos, cantidades)

plt.title("Producto más y menos vendido - Tienda 1")
plt.xlabel("Producto")
plt.ylabel("Cantidad de ventas")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Productos
productos = [
    ventas_tienda2.idxmax(), ventas_tienda2.idxmin(),
    ventas_tienda3.idxmax(), ventas_tienda3.idxmin(),
    ventas_tienda4.idxmax(), ventas_tienda4.idxmin()
]

# Cantidades
cantidades = [
    ventas_tienda2.max(), ventas_tienda2.min(),
    ventas_tienda3.max(), ventas_tienda3.min(),
    ventas_tienda4.max(), ventas_tienda4.min()
]

# Etiquetas
labels = [
    "T2 Más", "T2 Menos",
    "T3 Más", "T3 Menos",
    "T4 Más", "T4 Menos"
]

plt.figure()
plt.bar(labels, cantidades)

plt.title("Productos más y menos vendidos - Tiendas 2, 3 y 4")
plt.xlabel("Tiendas")
plt.ylabel("Cantidad de ventas")

plt.show()
```

# 5. Envío promedio por tienda

```python
envio_promedio_tienda1 = tienda["Costo de envío"].mean()
envio_promedio_tienda2 = tienda2["Costo de envío"].mean()
envio_promedio_tienda3 = tienda3["Costo de envío"].mean()
envio_promedio_tienda4 = tienda4["Costo de envío"].mean()

print("Costo de envío promedio Tienda 1:", envio_promedio_tienda1)
print("Costo de envío promedio Tienda 2:", envio_promedio_tienda2)
print("Costo de envío promedio Tienda 3:", envio_promedio_tienda3)
print("Costo de envío promedio Tienda 4:", envio_promedio_tienda4)
```

```python
import matplotlib.pyplot as plt

# Guardar los promedios en una lista
promedios_envio = [
    tienda["Costo de envío"].mean(),
    tienda2["Costo de envío"].mean(),
    tienda3["Costo de envío"].mean(),
    tienda4["Costo de envío"].mean()
]

# Nombres de las tiendas
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Crear gráfico
plt.figure()
plt.bar(tiendas, promedios_envio)

plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo de envío promedio")

plt.show()
```

Informe Final – Análisis de desempeño de tiendas

Introducción

El presente análisis tiene como objetivo evaluar el desempeño de las cuatro tiendas del Sr. Juan con el fin de determinar cuál de ellas sería la mejor candidata para vender. Para tomar esta decisión se analizaron distintos indicadores clave del negocio, tales como los ingresos totales generados por cada tienda, las categorías de productos más y menos vendidas, la calificación promedio otorgada por los clientes, los productos con mayor y menor volumen de ventas y el costo promedio de envío.

El análisis se realizó utilizando herramientas de análisis de datos en Python, específicamente con la librería pandas para el procesamiento de datos y matplotlib para la generación de visualizaciones.

Análisis de resultados
Ingresos totales por tienda

El análisis de los ingresos totales permitió identificar qué tiendas generan mayor volumen de ventas. Este indicador es uno de los más relevantes, ya que refleja directamente el desempeño comercial de cada tienda. A partir de los cálculos realizados, se observó que algunas tiendas presentan ingresos significativamente mayores, lo cual sugiere una mayor demanda de productos o una mejor estrategia de ventas.

Por el contrario, la tienda con menores ingresos evidencia un menor rendimiento en comparación con las demás.

Categorías de productos más vendidas

El análisis por categoría permitió identificar los tipos de productos con mayor demanda. Las categorías relacionadas con tecnología y artículos para el hogar, como electrónicos y electrodomésticos, tienden a concentrar una mayor proporción de ventas. Esto sugiere que estos productos tienen una fuerte aceptación entre los clientes.

Asimismo, se identificaron categorías con menor volumen de ventas, lo que puede indicar oportunidades para mejorar la oferta o replantear la estrategia comercial en esas áreas.

Calificación promedio de los clientes

La calificación promedio de los clientes permite evaluar el nivel de satisfacción con los productos y servicios ofrecidos por cada tienda. En general, las tiendas analizadas mantienen niveles de calificación relativamente altos, lo cual indica una experiencia de compra positiva.

Sin embargo, algunas tiendas presentan mejores calificaciones promedio que otras, lo que podría reflejar una mejor calidad en el servicio, logística o atención al cliente.

Productos más y menos vendidos

El análisis de los productos más vendidos permite identificar aquellos artículos que tienen mayor aceptación entre los consumidores. Estos productos representan oportunidades importantes para fortalecer la estrategia de ventas y aumentar la rentabilidad.

Por otro lado, los productos con menor volumen de ventas podrían requerir ajustes en estrategias de promoción, precio o posicionamiento en el mercado.

Costo promedio de envío

El costo promedio de envío es otro factor relevante que puede influir en la decisión de compra de los clientes. Aunque el envío es pagado por el cliente, costos más elevados pueden afectar la competitividad de la tienda frente a otras opciones del mercado.

Las tiendas con costos de envío más bajos pueden resultar más atractivas para los consumidores, mientras que costos más altos podrían representar una desventaja competitiva.

Conclusión y recomendación

Con base en el análisis de los diferentes indicadores evaluados, es posible identificar diferencias en el desempeño de las cuatro tiendas. Considerando factores como los ingresos generados, la satisfacción del cliente, el comportamiento de ventas por producto y los costos de envío, se puede determinar cuál tienda presenta un rendimiento inferior en comparación con las demás.

Por esta razón, se recomienda que el Sr. Juan considere vender la tienda que presenta menores ingresos y menor desempeño general, ya que su potencial de crecimiento y competitividad en el mercado parece ser más limitado.

Por otro lado, las tiendas que presentan mayores ingresos, mejores calificaciones de los clientes y una demanda más consistente de productos representan activos más sólidos y con mayor potencial de crecimiento a largo plazo.

Este análisis demuestra cómo el uso de datos y visualizaciones permite tomar decisiones estratégicas basadas en evidencia, facilitando una mejor gestión del negocio.
